```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
productname_vectorizer = TfidfVectorizer()
vectorised_product_names = productname_vectorizer.fit_transform(dframe.product_name).todense()
dframe['vectorised product names'] = vectorised_product_names.tolist()
dframe
```

```python
productdescr_vectorizer = TfidfVectorizer()
vectorised_product_description = productdescr_vectorizer.fit_transform(dframe.product_description).todense()
dframe['vectorised product descriptions'] = vectorised_product_description.tolist()
```

```python
location_vectorizer = TfidfVectorizer()
vectorised_location = location_vectorizer.fit_transform(dframe.location).todense()
dframe['vectorised location'] = vectorised_location.tolist()
```

```python
dframe
```

```python
X = dframe['vectorised product names'] + dframe['vectorised product descriptions'] + dframe['vectorised location']
```

```python
Y = dframe['price'].str.replace('£', '').str.replace(',', '').astype(float).values
Y
```

```python
import numpy as np
X = np.array(X.values.tolist())
```

```python
del dframe
del vectorised_product_description
del vectorised_location
del vectorised_product_names
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, Y)
```

```python
reg.score(X, Y)
```

```python
from PIL import Image
import os


def resize_image(final_size, im):
    size = im.size
    ratio = float(final_size) / max(size)
    new_image_size = tuple([int(x*ratio) for x in size])
    im = im.resize(new_image_size, Image.Resampling.LANCZOS)
    new_im = Image.new("RGB", (final_size, final_size))
    new_im.paste(im, ((final_size-new_image_size[0])//2, (final_size-new_image_size[1])//2))
    return new_im

if __name__ == '__main__':
    path = "/Users/macbook/Downloads/images/"
    dest_path = "/Users/macbook/Downloads/resized_images/"
    dirs = os.listdir(path)
    final_size = 400
    for n, item in enumerate(dirs[:5], 1):
        im = Image.open('/Users/macbook/Downloads/images/' + item)
        new_im = resize_image(final_size, im)
        new_im.save(dest_path + f'{n}_resized.jpg', 'JPEG', quality=100)
```
